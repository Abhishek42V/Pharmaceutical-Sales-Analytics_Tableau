This is an evaluation of the given dataset using python.


```python
#Importing the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
#reading the csv
df=pd.read_csv("DVDA1Salesdata.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#Check for null values
df.isnull().sum()
```

```python
#Quick Statistics on the numerical data
df.describe()
```

Given the above information, it is found that the dataframe attributes Sales2015,Sales2016,Units2015,Units2016 have values lesser than 0.

These values will be removed as they are considered outliers.

```python
"""
Removing the outliers from the data using
Sales data<0 and Units sold<0
"""
df_filtered=df[~((df["Sales2015"]<0)|(df["Sales2016"]<0)| (df["Units2015"]<0)| (df["Units2016"]<0))]
```

```python
#Cross checking the information pertaining to the existing dataframe
df_filtered.info()
```

Given the above, it is observed that a total of 35 entries were removed when the above filters were applied.

(Originally, there were 3709 values, post filtering there are 3674 values)

```python
#Quick statistics on the filtered dataframe
df_filtered.describe()
```

```python
#Exporting the filtered dataframe as a csv to import the same to tableau
df_filtered.to_csv('DVDA1_filtered.csv',index=False)
```
